```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. LOAD THE DATA
df = pd.read_csv('emails (1).csv')

```

```python
# 2. DATA CLEANING (As per EDA Lab 3)
# Check for null values
print("Missing values:\n", df.isnull().sum())
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Remove the "Subject: " prefix from the text column
df['text'] = df['text'].str.replace('Subject: ', '', regex=False)
```

```python
# 3. FEATURE ENGINEERING
# Create a 'length' column for message character count
df['length'] = df['text'].apply(len)
```

```python
# 4. VISUALIZATION (As per Matplotlib/Seaborn Labs 1 & 2)

# --- Plot 1: Class Balance (Seaborn Countplot) ---
plt.figure(figsize=(8, 6))
sns.countplot(x='spam', data=df, palette='viridis')
plt.title('Distribution of Spam (1) vs Ham (0)')
plt.xlabel('Email Type')
plt.ylabel('Number of Emails')
plt.show()
```

```python
# --- Plot 2: Distribution of Message Length (Seaborn Histplot) ---
plt.figure(figsize=(12, 6))
# We limit x-axis to 10,000 to see the distribution better (ignoring extreme outliers)
sns.histplot(data=df, x='length', hue='spam', bins=100, kde=True, palette='magma')
plt.title('Email Length Distribution by Category')
plt.xlim(0, 10000) 
plt.xlabel('Character Count')
plt.show()
```

```python
# --- Plot 3: Identifying Outliers (Seaborn Boxplot) ---
plt.figure(figsize=(10, 5))
sns.boxplot(x='spam', y='length', data=df)
plt.title('Boxplot of Message Length')
plt.show()
```
